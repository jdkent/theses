```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

## AIM 2:
Comparing the intranetwork correlations in betaseries correlations and residual correlations.

### Betaseries Correlations overall
Fronto-parietal correlations should be greater than Cingulo-Opercular correlations (overall).

```python
np.random.seed(123)
# beta series data (over, eg overall (not specific to condition))
betaseries_ovr = {}
betaseries_ovr['Fronto-Parietal'] = np.random.lognormal(0.4, sigma=0.1, size=50) - 1
betaseries_ovr['Cingulo-Opercular'] = np.random.lognormal(0.2, sigma=0.1, size=50) - 1

bsc_ovr_df = pd.DataFrame.from_dict(betaseries_ovr)
```

```python
sns.barplot(data=bsc_ovr_df, order=['Fronto-Parietal', 'Cingulo-Opercular'])
# label x axis
plt.xlabel('Networks', fontsize=14, weight='bold')
# label y axis
plt.ylabel("Pearson's R", fontsize=14, weight='bold')
# force y-axis to go from 0-1
plt.ylim(ymin=0, ymax=1)
plt.savefig('../img/Aim2/predict_betaseries_ovr.png')
```

### Betaseries Correlations (Incongruent - Congruent)
The fronto-parietal network difference between congruent and incongruent average connectivity should be greater than the congulo-opercular network difference between congruent and incongruent.

```python
np.random.seed(111)
# beta series data (spc = specific, eg condition specific)
betaseries_spc = {}
betaseries_spc['Fronto-Parietal'] = np.random.lognormal(0.2, sigma=0.1, size=50) - 1
betaseries_spc['Cingulo-Opercular'] = np.random.lognormal(0.05, sigma=0.1, size=50) - 1

bsc_spc_df = pd.DataFrame.from_dict(betaseries_spc)
```

```python
sns.barplot(data=bsc_spc_df, order=['Fronto-Parietal', 'Cingulo-Opercular'])
# label x axis
plt.xlabel('Networks', fontsize=14, weight='bold')
# label y axis
plt.ylabel("$\Delta$ BSC Pearson's R (Inc - Con)", fontsize=14, weight='bold')
# force y-axis to go from 0-1
plt.ylim(ymin=0, ymax=1)
plt.savefig('../img/Aim2/predict_betaseries_spc.png')
```

### Residual Correlations
Cingulo-Opercular Correlations should be greater than Fronto-Parietal Correlations

```python
np.random.seed(789)
# residual data
residuals = {}
residuals['Fronto-Parietal'] = np.random.lognormal(0.35, sigma=0.1, size=50) - 1
residuals['Cingulo-Opercular'] = np.random.lognormal(0.5, sigma=0.1, size=50) - 1

rc_df = pd.DataFrame.from_dict(residuals)
```

```python
sns.barplot(data=rc_df, order=['Fronto-Parietal', 'Cingulo-Opercular'])
# label x axis
plt.xlabel('Networks', fontsize=14, weight='bold')
# label y axis
plt.ylabel("Resid Pearson's R", fontsize=14, weight='bold')
# force y-axis to go from 0-1
plt.ylim(ymin=0, ymax=1)
plt.savefig('../img/Aim2/predict_residuals.png')
```

## AIM 3

- BetaSeries:
    - The Fronto-Parietal network will positively correlate with RT Cost, greater than the Cingulo-Opercular Network will correlate with RT cost.
    - The Inc-Con Fronto-Parietal Network will correlate with RT cost greater than the Inc-Con Cingular-Opercular network will correlate with RT cost.
    
- Residuals:
    - The Cingulo-Opercular Network will correlate with RT cost greater than the Fronto-Parietal Network.
    
(implicit: the variation in behavior explained by residuals should be different than the variance explained by betaseries)
